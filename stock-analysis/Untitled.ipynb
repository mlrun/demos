```python
import mlrun.feature_store as fs
import mlrun
import datetime
from mlrun.datastore.targets import ParquetTarget
import yfinance as yf
import pandas as pd
```

```python
mlrun.set_environment(project="stocks6-dani")
```

```python
start = datetime.datetime.now()-datetime.timedelta(5)
end = datetime.datetime.now()-datetime.timedelta(2)
```

```python
stocks_info_set = fs.FeatureSet("test_fs", entities=[fs.Entity("symbol")])
stocks_info_set.set_targets(targets=[ParquetTarget(partitioned=True,time_partitioning_granularity="day")],with_defaults=False)
```

```python
hist = yf.Ticker("GOOGL").history(start=start, end=end,interval = "1m").drop(["Dividends","Stock Splits"],axis=1)
fs_data = hist.copy().reset_index()
fs_data["symbol"]="GOOGL"
# fs_data.set_index("symbol",inplace=True)
```

```python
fs.ingest(stocks_info_set, fs_data, infer_options=fs.InferOptions.default())
```

```python
features = ["stocks.Opens_min_1h",
            "stocks.Opens_max_1h",
            "stocks.Volumes_min_1h",
            "stocks.Volumes_max_1h",
            "stocks.Open",
            "stocks.High",
            "stocks.Low",
            "stocks.Close",
            "stocks.Volume",
            "stocks.ticker",
            "news.Sentiment"]
vector = fs.FeatureVector("stocks-vec", features, description="stocks demo feature vector")
vector.save()
```

```python
last = datetime.datetime.now()-datetime.timedelta(8)
# change to read only data from the correct time and correct ticker
data = fs.get_offline_features(vector,entity_timestamp_column="Datetime",
                       start_time = last,
                       end_time = datetime.datetime.now()).to_dataframe()
```
